```python
import pandas as pd
from config_file import config
```

```python
annotations = pd.read_csv(config.DATA_PATH / "annotations.csv")
annotations.head()
```

```python
core_games = pd.read_csv(config.DATA_PATH / "core_games_info.csv")
core_games.head()
```

```python
core_annotations = annotations[
    annotations["filename"].apply(lambda x: x[:-4] in core_games["image_id"].array)
]
core_annotations.head()
```

```python
core_annotations.to_csv(config.DATA_PATH / "core_annotations.csv", index=False)
```
